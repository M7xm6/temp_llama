```python
"""
Llama2 对抗性后缀优化 - REINFORCE 版
直接优化前100步平均EOS logit，无需反向传播通过生成步骤
"""

import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import AutoModelForCausalLM, AutoTokenizer
import numpy as np
from tqdm import tqdm
import gc

class SuffixOptimizerREINFORCE:
    def __init__(self, model, tokenizer, prompt, suffix_init="########",
                 N=100, T=80, batch_size=64, lr=0.01, temperature=1.0):
        """
        参数:
            N: 生成步数（用于计算奖励，建议设为100以匹配你的分析）
            T: 优化迭代次数
            batch_size: 每轮采样的后缀数量
            lr: 学习率
            temperature: 采样温度（>1 增强探索，<1 增强利用）
        """
        self.model = model
        self.tokenizer = tokenizer
        self.device = model.device

        # Tokenize prompt
        self.prompt_tokens = tokenizer.encode(prompt, add_special_tokens=True,
                                               return_tensors='pt').to(self.device)

        # 初始后缀 tokens
        suffix_text = suffix_init
        init_suffix_tokens = tokenizer.encode(suffix_text, add_special_tokens=False,
                                               return_tensors='pt').to(self.device)
        self.suffix_len = init_suffix_tokens.shape[1]

        self.N = N
        self.T = T
        self.batch_size = batch_size
        self.lr = lr
        self.temperature = temperature

        self.vocab_size = model.config.vocab_size
        self.eos_token_id = tokenizer.eos_token_id

        self.loss_history = []  # 存储每轮的平均奖励
        self.best_suffix = init_suffix_tokens.clone()  # 当前最佳后缀
        self.best_reward = -float('inf')

        # ===== 可学习的策略参数 =====
        # logits 形状 [suffix_len, vocab_size]
        self.logits = nn.Parameter(torch.randn(self.suffix_len, self.vocab_size,
                                                device=self.device) * 0.01)
        # 初始化：让初始后缀对应的 token 有稍高的 logit
        for pos, token_id in enumerate(init_suffix_tokens[0]):
            self.logits.data[pos, token_id] += 1.0

        self.optimizer = torch.optim.Adam([self.logits], lr=self.lr)

        print(f"初始化完成:")
        print(f"  Prompt tokens: {self.prompt_tokens.shape}")
        print(f"  Suffix length: {self.suffix_len}")
        print(f"  N (生成步数): {self.N}")
        print(f"  T (迭代次数): {self.T}")
        print(f"  Batch size: {self.batch_size}")
        print(f"  学习率: {self.lr}")
        print(f"  温度: {self.temperature}")

    def compute_eos_reward_batch(self, suffix_tokens_batch):
        """
        批量计算奖励 = 前 N 步的平均 EOS logit
        suffix_tokens_batch: [batch_size, suffix_len]
        返回: [batch_size] 的奖励值（越大越好）
        """
        batch_size = suffix_tokens_batch.shape[0]

        # 扩展prompt到batch
        prompt_batch = self.prompt_tokens.repeat(batch_size, 1)  # [batch_size, prompt_len]

        # 拼接
        input_ids = torch.cat([prompt_batch, suffix_tokens_batch], dim=1)  # [batch_size, total_len]

        rewards = []  # 存储每个样本的累积 EOS logit

        with torch.no_grad():
            current_ids = input_ids

            for step in range(self.N):
                # 前向传播
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]  # [batch_size, vocab_size]

                # 获取EOS logit
                eos_logits = logits[:, self.eos_token_id]  # [batch_size]
                rewards.append(eos_logits)

                # 屏蔽EOS，贪婪选择下一个token（用于继续生成）
                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_tokens = torch.argmax(masked_logits, dim=-1, keepdim=True)  # [batch_size, 1]

                # 更新输入
                current_ids = torch.cat([current_ids, next_tokens], dim=1)

        # 计算平均EOS logit 作为奖励
        rewards = torch.stack(rewards, dim=1)  # [batch_size, N]
        avg_rewards = rewards.mean(dim=1)      # [batch_size]
        return -avg_rewards

    def sample_suffixes(self, batch_size):
        """从当前策略采样一批后缀"""
        # 应用温度
        scaled_logits = self.logits / self.temperature
        probs = F.softmax(scaled_logits, dim=-1)  # [suffix_len, vocab_size]

        # 对每个位置独立采样
        # 返回形状: [suffix_len, batch_size]
        sampled_indices = torch.multinomial(probs, num_samples=batch_size, replacement=True)
        # 转置为 [batch_size, suffix_len]
        sampled_suffixes = sampled_indices.T
        return sampled_suffixes, probs

    def compute_log_probs(self, sampled_suffixes, probs):
        """
        计算每个采样后缀的对数概率
        sampled_suffixes: [batch_size, suffix_len]
        probs: [suffix_len, vocab_size]
        返回: [batch_size]
        """
        batch_size = sampled_suffixes.shape[0]
        suffix_len = self.suffix_len

        # 扩展 probs 以便 gather
        # probs_expanded: [suffix_len, batch_size, vocab_size]
        probs_expanded = probs.unsqueeze(1).expand(-1, batch_size, -1)

        # 获取每个位置选中token的概率
        # token_probs: [suffix_len, batch_size]
        token_probs = probs_expanded.gather(2, sampled_suffixes.T.unsqueeze(-1)).squeeze(-1)

        # 求和得到对数概率（加小常数防止log(0)）
        log_probs = torch.log(token_probs + 1e-10).sum(dim=0)  # [batch_size]
        return log_probs

    def optimize(self):
        """主优化循环 - REINFORCE"""
        print(f"\n开始 REINFORCE 优化 (T={self.T}次迭代, 奖励基于{self.N}步平均EOS)")
        print("=" * 60)

        for iteration in tqdm(range(self.T), desc="优化进度"):
            # 定期清理显存
            if iteration % 10 == 0:
                torch.cuda.empty_cache()
                gc.collect()

            # 1. 从当前策略采样一批后缀
            sampled_suffixes, probs = self.sample_suffixes(self.batch_size)

            # 2. 计算奖励（无梯度）
            with torch.no_grad():
                rewards = self.compute_eos_reward_batch(sampled_suffixes)  # [batch_size]

            # 3. 计算对数概率
            log_probs = self.compute_log_probs(sampled_suffixes, probs)

            # 4. 基线（使用 batch 内平均奖励）
            baseline = rewards.mean()
            advantages = rewards - baseline

            # 5. REINFORCE 损失（最大化期望奖励）
            loss = -(advantages.detach() * log_probs).mean()

            # 6. 反向传播更新策略
            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()

            # 7. 记录最佳后缀
            batch_best_reward = rewards.max().item()
            if batch_best_reward > self.best_reward:
                self.best_reward = batch_best_reward
                best_idx = torch.argmax(rewards)
                self.best_suffix = sampled_suffixes[best_idx:best_idx+1]

            # 记录平均奖励
            avg_reward = rewards.mean().item()
            self.loss_history.append(avg_reward)

            # 打印进度
            if (iteration + 1) % 10 == 0 or iteration == 0:
                suffix_text = self.tokenizer.decode(self.best_suffix[0], skip_special_tokens=True)
                print(f"\nIter {iteration+1}/{self.T}:")
                print(f"  平均奖励: {avg_reward:.4f}")
                print(f"  最佳奖励: {self.best_reward:.4f}")
                print(f"  当前最佳后缀: '{suffix_text}'")
                if torch.cuda.is_available():
                    print(f"  GPU内存: {torch.cuda.memory_allocated()/1e9:.2f}GB")

        print("\n" + "=" * 60)
        print("优化完成!")
        return self.best_suffix

    def test_generation(self, suffix_tokens, max_new_tokens=800):
        """测试生成效果"""
        print("\n" + "=" * 60)
        print("测试生成效果:")
        print("=" * 60)

        input_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)

        with torch.no_grad():
            outputs = self.model.generate(
                input_ids,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=self.tokenizer.eos_token_id,
                use_cache=True
            )

        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_tokens = outputs.shape[1] - input_ids.shape[1]

        print(f"\n生成文本:\n{generated_text}\n")
        print(f"生成token数: {generated_tokens}")

        return generated_text, generated_tokens


# ==================== 主程序 ====================
if __name__ == "__main__":
    # GPU检查
    print("=" * 60)
    print("系统检查:")
    print("=" * 60)
    print(f"PyTorch版本: {torch.__version__}")
    print(f"CUDA是否可用: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"CUDA版本: {torch.version.cuda}")
        print(f"GPU设备: {torch.cuda.get_device_name(0)}")
        print(f"GPU内存: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True
    else:
        print("⚠️ 警告: 未检测到GPU")

    print("\n加载模型...")

    # 加载模型（请根据实际情况修改路径或使用缓存）
    model_name = "meta-llama/Llama-2-7b-chat-hf"

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token

    # 4-bit量化配置（减少显存占用）
    from transformers import BitsAndBytesConfig

    quantization_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4"
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=quantization_config,
        device_map="auto",
        torch_dtype=torch.float16,
        low_cpu_mem_usage=True
    )

    model.eval()
    print(f"模型加载完成，设备: {model.device}")

    # 测试问题
    prompt = "Please explain quantum computing in simple terms."

    # 基线测试
    print("\n" + "=" * 60)
    print("基线测试 (无后缀):")
    print("=" * 60)

    baseline_input = tokenizer.encode(prompt, return_tensors='pt').to(model.device)
    with torch.no_grad():
        baseline_output = model.generate(
            baseline_input,
            max_new_tokens=800,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
            use_cache=True
        )
    baseline_text = tokenizer.decode(baseline_output[0], skip_special_tokens=True)
    baseline_length = baseline_output.shape[1] - baseline_input.shape[1]

    print(f"\n基线生成:\n{baseline_text}\n")
    print(f"基线生成token数: {baseline_length}")

    # 清理显存
    del baseline_output
    torch.cuda.empty_cache()
    gc.collect()

    # 开始 REINFORCE 优化
    print("\n" + "=" * 60)
    print("开始后缀优化 (REINFORCE):")
    print("=" * 60)

    optimizer = SuffixOptimizerREINFORCE(
        model=model,
        tokenizer=tokenizer,
        prompt=prompt,
        suffix_init="########",
        N=100,           # 使用前100步平均EOS作为奖励
        T=80,            # 迭代次数
        batch_size=64,   # 每批采样64个后缀
        lr=0.01,         # 学习率
        temperature=1.0  # 初始温度
    )

    # 运行优化
    optimized_suffix = optimizer.optimize()

    # 清理显存
    torch.cuda.empty_cache()
    gc.collect()

    # 测试优化后的后缀
    optimized_text, optimized_length = optimizer.test_generation(optimized_suffix)

    # 对比
    print("\n" + "=" * 60)
    print("最终对比:")
    print("=" * 60)
    print(f"基线生成token数: {baseline_length}")
    print(f"优化后生成token数: {optimized_length}")
    print(f"提升: {optimized_length - baseline_length} tokens ({(optimized_length/baseline_length - 1)*100:.1f}%)")
    print(f"\n优化后的后缀: '{tokenizer.decode(optimized_suffix[0], skip_special_tokens=True)}'")

    # 保存奖励曲线
    if len(optimizer.loss_history) > 0:
        import matplotlib.pyplot as plt

        plt.figure(figsize=(10, 5))
        plt.plot(optimizer.loss_history)
        plt.xlabel('Iteration')
        plt.ylabel('Average Reward (EOS logit)')
        plt.title('REINFORCE Optimization Progress')
        plt.grid(True, alpha=0.3)
        plt.savefig('reinforce_reward_curve.png', dpi=150, bbox_inches='tight')
        print("\n奖励曲线已保存至 reinforce_reward_curve.png")
```
