```python
"""
Llama2 对抗性后缀优化 - 修复版
解决显存爆炸和效率问题
"""

import torch
import torch.nn.functional as F
from transformers import AutoModelForCausalLM, AutoTokenizer
import numpy as np
from tqdm import tqdm
import gc

class SuffixOptimizerFixed:
    def __init__(self, model, tokenizer, prompt, suffix_init="########", N=20, T=80, k=256, batch_size=64):
        """
        参数优化:
        - batch_size降低到64（原512太大）
        - 添加显存管理
        """
        self.model = model
        self.tokenizer = tokenizer
        self.device = model.device

        # Tokenize
        self.prompt_tokens = tokenizer.encode(prompt, add_special_tokens=True, return_tensors='pt').to(self.device)
        suffix_text = suffix_init
        self.suffix_tokens = tokenizer.encode(suffix_text, add_special_tokens=False, return_tensors='pt').to(self.device)

        self.N = N
        self.T = T
        self.k = k
        self.batch_size = batch_size

        self.vocab_size = model.config.vocab_size
        self.eos_token_id = tokenizer.eos_token_id

        self.loss_history = []

        # 检查模型dtype
        embedding_dtype = model.get_input_embeddings().weight.dtype

        print(f"初始化完成:")
        print(f"  Prompt: {prompt}")
        print(f"  初始后缀: {suffix_init}")
        print(f"  Prompt tokens: {self.prompt_tokens.shape}")
        print(f"  Suffix tokens: {self.suffix_tokens.shape}")
        print(f"  EOS token ID: {self.eos_token_id}")
        print(f"  Vocab size: {self.vocab_size}")
        print(f"  Embedding dtype: {embedding_dtype}")

    def compute_eos_loss_batch(self, suffix_tokens_batch):
        """
        批量计算EOS损失（无梯度，高效）
        suffix_tokens_batch: [batch_size, suffix_len]
        返回: [batch_size] 的损失值
        """
        batch_size = suffix_tokens_batch.shape[0]

        # 扩展prompt到batch
        prompt_batch = self.prompt_tokens.repeat(batch_size, 1)  # [batch_size, prompt_len]

        # 拼接
        input_ids = torch.cat([prompt_batch, suffix_tokens_batch], dim=1)  # [batch_size, total_len]

        losses = []

        with torch.no_grad():
            current_ids = input_ids

            for step in range(self.N):
                # 前向传播
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]  # [batch_size, vocab_size]

                # 获取EOS logit
                eos_logits = logits[:, self.eos_token_id]  # [batch_size]
                losses.append(eos_logits)

                # 屏蔽EOS，继续生成
                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_tokens = torch.argmax(masked_logits, dim=-1, keepdim=True)  # [batch_size, 1]

                # 更新输入
                current_ids = torch.cat([current_ids, next_tokens], dim=1)

        # 计算平均EOS logit
        losses = torch.stack(losses, dim=1)  # [batch_size, N]
        avg_losses = losses.mean(dim=1)  # [batch_size]

        return avg_losses

    def compute_loss_with_grad_FIXED(self, suffix_tokens):
        """
        修复版梯度计算：针对量化模型优化
        """
        # 关键修复1：直接将one-hot转为embedding的dtype
        embed_weights = self.model.get_input_embeddings().weight
        target_dtype = embed_weights.dtype

        # One-hot relaxation并立即转换为目标dtype
        suffix_onehot = F.one_hot(suffix_tokens[0], num_classes=self.vocab_size).to(target_dtype)
        suffix_onehot.requires_grad_(True)

        # 计算suffix embeddings
        suffix_embeds = torch.matmul(suffix_onehot, embed_weights).unsqueeze(0)  # [1, suffix_len, hidden_dim]

        # 获取prompt embeddings
        with torch.no_grad():
            prompt_embeds = self.model.get_input_embeddings()(self.prompt_tokens)

        # 确保两者dtype一致
        if suffix_embeds.dtype != prompt_embeds.dtype:
            suffix_embeds = suffix_embeds.to(prompt_embeds.dtype)

        # ===== 关键修复：只计算第一步的梯度 =====
        input_embeds = torch.cat([prompt_embeds, suffix_embeds], dim=1)

        # 第一步：计算梯度
        outputs = self.model(inputs_embeds=input_embeds, use_cache=False)
        first_logits = outputs.logits[:, -1, :]
        first_eos_logit = first_logits[0, self.eos_token_id]

        # 继续生成N-1步（无梯度）
        with torch.no_grad():
            current_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)
            current_ids = torch.cat([current_ids, torch.argmax(first_logits, dim=-1, keepdim=True)], dim=1)

            total_eos_logit = first_eos_logit.item()

            for step in range(1, self.N):
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]

                eos_logit = logits[0, self.eos_token_id]
                total_eos_logit += eos_logit.item()

                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_token = torch.argmax(masked_logits, dim=-1, keepdim=True)

                current_ids = torch.cat([current_ids, next_token], dim=1)

        # 只对第一步的EOS logit计算梯度（代理整体趋势）
        loss = first_eos_logit
        loss.backward()

        gradients = suffix_onehot.grad.unsqueeze(0)  # [1, suffix_len, vocab_size]

        avg_loss = total_eos_logit / self.N

        return avg_loss, gradients

    def get_top_k_candidates(self, gradients, suffix_tokens):
        """
        为每个位置选择top-k候选
        """
        suffix_len = suffix_tokens.shape[1]
        candidates = {}

        for pos in range(suffix_len):
            grad = gradients[0, pos, :]
            top_indices = torch.topk(-grad, self.k).indices
            candidates[pos] = top_indices.cpu().tolist()

        return candidates

    def optimize(self):
        """
        主优化循环 - 批量评估版本
        """
        print(f"\n开始优化 (T={self.T}次迭代, N={self.N}步生成)")
        print("=" * 60)

        current_suffix = self.suffix_tokens.clone()

        for iteration in tqdm(range(self.T), desc="优化进度"):
            # ===== 修改1：添加显存清理 =====
            if iteration % 10 == 0:
                torch.cuda.empty_cache()
                gc.collect()

            # 1. 计算梯度
            try:
                loss, gradients = self.compute_loss_with_grad_FIXED(current_suffix)
                self.loss_history.append(loss)
            except RuntimeError as e:
                print(f"\n⚠️ 梯度计算失败: {e}")
                print("尝试降低batch_size或N")
                break

            # 2. 获取候选
            candidates = self.get_top_k_candidates(gradients, current_suffix)

            # ===== 修改2：批量评估候选（大幅提速）=====
            suffix_len = current_suffix.shape[1]

            # 生成候选后缀（改为批量）
            candidate_batch = []
            for _ in range(self.batch_size):
                pos = np.random.randint(0, suffix_len)
                new_token = np.random.choice(candidates[pos])

                new_suffix = current_suffix.clone()
                new_suffix[0, pos] = new_token
                candidate_batch.append(new_suffix)

            # 拼接成batch
            candidate_batch = torch.cat(candidate_batch, dim=0)  # [batch_size, suffix_len]

            # 批量评估
            try:
                candidate_losses = self.compute_eos_loss_batch(candidate_batch)
            except RuntimeError as e:
                print(f"\n⚠️ 批量评估失败: {e}")
                print("降低batch_size")
                self.batch_size = max(8, self.batch_size // 2)
                continue

            # 选择最佳
            best_idx = torch.argmin(candidate_losses).item()
            best_loss = candidate_losses[best_idx].item()

            if best_loss < loss:
                current_suffix = candidate_batch[best_idx:best_idx+1]

            # 3. 打印进度
            if (iteration + 1) % 10 == 0 or iteration == 0:
                suffix_text = self.tokenizer.decode(current_suffix[0], skip_special_tokens=True)
                print(f"\nIter {iteration+1}/{self.T}:")
                print(f"  损失: {min(best_loss, loss):.4f}")
                print(f"  当前后缀: '{suffix_text}'")
                if torch.cuda.is_available():
                    print(f"  GPU内存: {torch.cuda.memory_allocated()/1e9:.2f}GB")

        print("\n" + "=" * 60)
        print("优化完成!")

        return current_suffix

    def test_generation(self, suffix_tokens, max_new_tokens=800):
        """测试生成"""
        print("\n" + "=" * 60)
        print("测试生成效果:")
        print("=" * 60)

        input_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)

        with torch.no_grad():
            outputs = self.model.generate(
                input_ids,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=self.tokenizer.eos_token_id,
                use_cache=True
            )

        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_tokens = outputs.shape[1] - input_ids.shape[1]

        print(f"\n生成文本:\n{generated_text}\n")
        print(f"生成token数: {generated_tokens}")

        return generated_text, generated_tokens


# ==================== 主程序 ====================
if __name__ == "__main__":
    # GPU检查
    print("=" * 60)
    print("系统检查:")
    print("=" * 60)
    print(f"PyTorch版本: {torch.__version__}")
    print(f"CUDA是否可用: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"CUDA版本: {torch.version.cuda}")
        print(f"GPU设备: {torch.cuda.get_device_name(0)}")
        print(f"GPU内存: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

        # 设置显存优化
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True
    else:
        print("⚠️ 警告: 未检测到GPU")

    print("\n加载模型...")

    # 加载模型
    model_name = "meta-llama/Llama-2-7b-chat-hf"

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token

    # 4-bit量化
    from transformers import BitsAndBytesConfig

    quantization_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4"
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=quantization_config,
        device_map="auto",
        torch_dtype=torch.float16,
        low_cpu_mem_usage=True
    )

    # 设置为eval模式
    model.eval()

    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print(f"模型加载完成")
    print(f"模型设备: {model.device}")
    if torch.cuda.is_available():
        print(f"当前GPU内存使用: {torch.cuda.memory_allocated(0) / 1e9:.2f} GB")

    # 测试问题
    prompt = "Please explain quantum computing in simple terms."

    # 基线测试
    print("\n" + "=" * 60)
    print("基线测试 (无后缀):")
    print("=" * 60)

    baseline_input = tokenizer.encode(prompt, return_tensors='pt').to(model.device)
    with torch.no_grad():
        baseline_output = model.generate(
            baseline_input,
            max_new_tokens=800,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
            use_cache=True
        )
    baseline_text = tokenizer.decode(baseline_output[0], skip_special_tokens=True)
    baseline_length = baseline_output.shape[1] - baseline_input.shape[1]

    print(f"\n基线生成:\n{baseline_text}\n")
    print(f"基线生成token数: {baseline_length}")

    # 清理显存
    del baseline_output
    torch.cuda.empty_cache()
    gc.collect()

    # 开始优化
    print("\n" + "=" * 60)
    print("开始后缀优化:")
    print("=" * 60)

    optimizer = SuffixOptimizerFixed(
        model=model,
        tokenizer=tokenizer,
        prompt=prompt,
        suffix_init="########",
        N=20,
        T=80,
        k=256,
        batch_size=64  # 降低到64
    )

    # 运行优化
    optimized_suffix = optimizer.optimize()

    # 清理显存
    torch.cuda.empty_cache()
    gc.collect()

    # 测试结果
    optimized_text, optimized_length = optimizer.test_generation(optimized_suffix)

    # 对比
    print("\n" + "=" * 60)
    print("最终对比:")
    print("=" * 60)
    print(f"基线生成token数: {baseline_length}")
    print(f"优化后生成token数: {optimized_length}")
    print(f"提升: {optimized_length - baseline_length} tokens ({(optimized_length/baseline_length - 1)*100:.1f}%)")
    print(f"\n优化后的后缀: '{tokenizer.decode(optimized_suffix[0], skip_special_tokens=True)}'")

    # 保存损失曲线
    if len(optimizer.loss_history) > 0:
        import matplotlib.pyplot as plt

        plt.figure(figsize=(10, 5))
        plt.plot(optimizer.loss_history)
        plt.xlabel('Iteration')
        plt.ylabel('Average EOS Logit')
        plt.title('EOS Loss Optimization Progress')
        plt.grid(True, alpha=0.3)
        plt.savefig('loss_curve.png', dpi=150, bbox_inches='tight')
        print("\n损失曲线已保存至 loss_curve.png")
```

```python
import torch
import torch.nn.functional as F
from transformers import AutoModelForCausalLM, AutoTokenizer
import numpy as np
from tqdm import tqdm
import gc
import matplotlib.pyplot as plt

class GCGStyleOptimizer:
    def __init__(self, model, tokenizer, prompt, suffix_init="########",
                 N=100, T=80, k=128, grad_batch_size=1, eval_batch_size=32):
        """
        参数:
            N: 生成步数（用于计算平均 EOS logit）
            T: 优化迭代次数
            k: 每个位置保留的候选 token 数量
            grad_batch_size: 梯度计算时的 batch size（当前版本固定为1）
            eval_batch_size: 评估候选时的批次大小（控制显存）
        """
        self.model = model
        self.tokenizer = tokenizer
        self.device = model.device

        # 编码 prompt 和初始后缀
        self.prompt_tokens = tokenizer.encode(prompt, add_special_tokens=True,
                                               return_tensors='pt').to(self.device)
        suffix_text = suffix_init
        self.suffix_tokens = tokenizer.encode(suffix_text, add_special_tokens=False,
                                               return_tensors='pt').to(self.device)

        self.N = N
        self.T = T
        self.k = k
        self.eval_batch_size = eval_batch_size

        self.vocab_size = model.config.vocab_size
        self.eos_token_id = tokenizer.eos_token_id

        self.loss_history = []

        # 检查模型 dtype
        embedding_dtype = model.get_input_embeddings().weight.dtype
        print(f"初始化完成:")
        print(f"  Prompt tokens: {self.prompt_tokens.shape}")
        print(f"  Suffix tokens: {self.suffix_tokens.shape}")
        print(f"  EOS token ID: {self.eos_token_id}")
        print(f"  Vocab size: {self.vocab_size}")
        print(f"  Embedding dtype: {embedding_dtype}")
        print(f"  N={N}, T={T}, k={k}, eval_batch_size={eval_batch_size}")

    def compute_eos_loss_batch(self, suffix_tokens_batch):
        """
        批量计算平均 EOS logit（无梯度，高效）
        suffix_tokens_batch: [batch_size, suffix_len]
        返回: [batch_size] 的平均 EOS logit（越小越好）
        """
        batch_size = suffix_tokens_batch.shape[0]
        prompt_batch = self.prompt_tokens.repeat(batch_size, 1)  # [batch_size, prompt_len]
        input_ids = torch.cat([prompt_batch, suffix_tokens_batch], dim=1)

        losses = []
        with torch.no_grad():
            current_ids = input_ids
            for step in range(self.N):
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]                # [batch_size, vocab_size]
                eos_logits = logits[:, self.eos_token_id]        # [batch_size]
                losses.append(eos_logits)

                # 贪婪生成下一步（屏蔽 EOS）
                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_tokens = torch.argmax(masked_logits, dim=-1, keepdim=True)
                current_ids = torch.cat([current_ids, next_tokens], dim=1)

        losses = torch.stack(losses, dim=1)                     # [batch_size, N]
        avg_losses = losses.mean(dim=1)                          # [batch_size]
        return avg_losses

    def compute_loss_with_grad(self, suffix_tokens):
        """
        计算梯度（基于第一步的 EOS logit，作为整体趋势的代理）
        suffix_tokens: [1, suffix_len]
        返回: (avg_eos_logit, gradients)  gradients: [1, suffix_len, vocab_size]
        """
        embed_weights = self.model.get_input_embeddings().weight
        target_dtype = embed_weights.dtype

        # One-hot relaxation
        suffix_onehot = F.one_hot(suffix_tokens[0], num_classes=self.vocab_size).to(target_dtype)
        suffix_onehot.requires_grad_(True)

        # 计算 suffix embeddings
        suffix_embeds = torch.matmul(suffix_onehot, embed_weights).unsqueeze(0)  # [1, suffix_len, hidden]

        # 获取 prompt embeddings
        with torch.no_grad():
            prompt_embeds = self.model.get_input_embeddings()(self.prompt_tokens)

        # 确保 dtype 一致
        if suffix_embeds.dtype != prompt_embeds.dtype:
            suffix_embeds = suffix_embeds.to(prompt_embeds.dtype)

        # 拼接
        input_embeds = torch.cat([prompt_embeds, suffix_embeds], dim=1)

        # 第一步前向传播
        outputs = self.model(inputs_embeds=input_embeds, use_cache=False)
        first_logits = outputs.logits[:, -1, :]                 # [1, vocab_size]
        first_eos_logit = first_logits[0, self.eos_token_id]

        # 继续生成 N-1 步（无梯度），累加 EOS logit
        with torch.no_grad():
            current_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)
            current_ids = torch.cat([current_ids, torch.argmax(first_logits, dim=-1, keepdim=True)], dim=1)

            total_eos_logit = first_eos_logit.item()

            for step in range(1, self.N):
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]
                eos_logit = logits[0, self.eos_token_id]
                total_eos_logit += eos_logit.item()

                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_token = torch.argmax(masked_logits, dim=-1, keepdim=True)
                current_ids = torch.cat([current_ids, next_token], dim=1)

        # 仅对第一步的 EOS logit 反向传播
        loss = first_eos_logit
        loss.backward()

        gradients = suffix_onehot.grad.unsqueeze(0)              # [1, suffix_len, vocab_size]
        avg_loss = total_eos_logit / self.N

        return avg_loss, gradients

    def get_top_k_candidates(self, gradients, suffix_tokens):
        """
        为每个位置选取 top-k 候选 token（基于负梯度）
        """
        suffix_len = suffix_tokens.shape[1]
        candidates = {}
        for pos in range(suffix_len):
            grad = gradients[0, pos, :]                         # [vocab_size]
            # 负梯度越大，替换后越可能降低损失
            top_indices = torch.topk(-grad, self.k).indices
            candidates[pos] = top_indices.cpu().tolist()
        return candidates

    def optimize(self):
        print(f"\n开始 GCG 风格优化 (T={self.T}次迭代, N={self.N}步生成, k={self.k})")
        print("=" * 70)

        current_suffix = self.suffix_tokens.clone()

        for iteration in tqdm(range(self.T), desc="优化进度"):
            # 定期清理显存
            if iteration % 10 == 0:
                torch.cuda.empty_cache()
                gc.collect()

            # 1. 计算梯度，获取候选
            loss, gradients = self.compute_loss_with_grad(current_suffix)
            candidates = self.get_top_k_candidates(gradients, current_suffix)

            suffix_len = current_suffix.shape[1]
            best_suffix = current_suffix.clone()
            best_loss = loss

            # 2. 遍历每个位置及其候选，寻找最佳替换
            for pos in range(suffix_len):
                pos_candidates = candidates[pos]                 # list of token ids
                num_cand = len(pos_candidates)

                # 分批评估，避免显存溢出
                for i in range(0, num_cand, self.eval_batch_size):
                    batch_cand = pos_candidates[i:i+self.eval_batch_size]
                    batch_size_curr = len(batch_cand)

                    # 构造 batch 后缀
                    cand_suffixes = current_suffix.repeat(batch_size_curr, 1)  # [batch_size_curr, suffix_len]
                    cand_suffixes[:, pos] = torch.tensor(batch_cand, device=self.device)

                    # 评估这批候选
                    cand_losses = self.compute_eos_loss_batch(cand_suffixes)   # [batch_size_curr]

                    # 检查是否有更优的
                    min_idx = torch.argmin(cand_losses).item()
                    if cand_losses[min_idx] < best_loss:
                        best_loss = cand_losses[min_idx].item()
                        best_suffix = cand_suffixes[min_idx:min_idx+1]

            # 3. 更新当前后缀
            current_suffix = best_suffix
            self.loss_history.append(best_loss)

            # 4. 打印进度
            if (iteration + 1) % 10 == 0 or iteration == 0:
                suffix_text = self.tokenizer.decode(current_suffix[0], skip_special_tokens=True)
                print(f"\nIter {iteration+1}/{self.T}:")
                print(f"  损失 (平均EOS logit): {best_loss:.4f}")
                print(f"  当前后缀: '{suffix_text}'")
                if torch.cuda.is_available():
                    print(f"  GPU内存: {torch.cuda.memory_allocated()/1e9:.2f}GB")

        print("\n" + "=" * 70)
        print("优化完成!")
        return current_suffix

    def test_generation(self, suffix_tokens, max_new_tokens=2000):
        """测试生成效果（贪婪解码）"""
        print("\n" + "=" * 70)
        print("测试生成效果:")
        print("=" * 70)

        input_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)

        with torch.no_grad():
            outputs = self.model.generate(
                input_ids,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=self.tokenizer.eos_token_id,
                use_cache=True
            )

        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_tokens = outputs.shape[1] - input_ids.shape[1]

        print(f"\n生成文本:\n{generated_text}\n")
        print(f"生成token数: {generated_tokens}")

        return generated_text, generated_tokens


if __name__ == "__main__":
    # GPU检查
    print("=" * 60)
    print("系统检查:")
    print("=" * 60)
    print(f"PyTorch版本: {torch.__version__}")
    print(f"CUDA是否可用: {torch.cuda.is_available()}")
    if torch.cuda.is_available():
        print(f"CUDA版本: {torch.version.cuda}")
        print(f"GPU设备: {torch.cuda.get_device_name(0)}")
        print(f"GPU内存: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")

        # 设置显存优化
        torch.cuda.empty_cache()
        torch.backends.cudnn.benchmark = True
    else:
        print("⚠️ 警告: 未检测到GPU")

    print("\n加载模型...")

    # 加载模型
    model_name = "meta-llama/Llama-2-7b-chat-hf"

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token

    # 4-bit量化
    from transformers import BitsAndBytesConfig

    quantization_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
        bnb_4bit_quant_type="nf4"
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        quantization_config=quantization_config,
        device_map="auto",
        torch_dtype=torch.float16,
        low_cpu_mem_usage=True
    )

    # 设置为eval模式
    model.eval()

    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print(f"模型加载完成")
    print(f"模型设备: {model.device}")
    if torch.cuda.is_available():
        print(f"当前GPU内存使用: {torch.cuda.memory_allocated(0) / 1e9:.2f} GB")


    prompt = "Please explain quantum computing in simple terms."

    # 基线测试（无后缀）
    baseline_input = tokenizer.encode(prompt, return_tensors='pt').to(model.device)
    with torch.no_grad():
        baseline_output = model.generate(
            baseline_input,
            max_new_tokens=2000,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
            use_cache=True
        )
    baseline_length = baseline_output.shape[1] - baseline_input.shape[1]
    print(f"基线生成token数: {baseline_length}")

    # 初始化优化器
    optimizer = GCGStyleOptimizer(
        model=model,
        tokenizer=tokenizer,
        prompt=prompt,
        suffix_init="########",
        N=100,               # 前100步的平均EOS logit
        T=50,                # 迭代次数
        k=64,               # 每个位置保留128个候选
        eval_batch_size=16   # 评估时的批次大小，根据显存调整
    )

    # 运行优化
    optimized_suffix = optimizer.optimize()

    # 测试优化后的后缀
    optimized_text, optimized_length = optimizer.test_generation(optimized_suffix)

    # 对比结果
    print("\n" + "=" * 70)
    print("最终对比:")
    print("=" * 70)
    print(f"基线生成token数: {baseline_length}")
    print(f"优化后生成token数: {optimized_length}")
    print(f"提升: {optimized_length - baseline_length} tokens ({(optimized_length/baseline_length - 1)*100:.1f}%)")
    print(f"\n优化后的后缀: '{tokenizer.decode(optimized_suffix[0], skip_special_tokens=True)}'")

    # 绘制损失曲线
    if len(optimizer.loss_history) > 0:
        plt.figure(figsize=(10, 5))
        plt.plot(optimizer.loss_history)
        plt.xlabel('Iteration')
        plt.ylabel('Average EOS Logit')
        plt.title('GCG-Style Optimization Progress')
        plt.grid(True, alpha=0.3)
        plt.savefig('gcg_loss_curve.png', dpi=150, bbox_inches='tight')
        print("\n损失曲线已保存至 gcg_loss_curve.png")
```

```python
import torch
import torch.nn.functional as F
from transformers import AutoModelForCausalLM, AutoTokenizer
import numpy as np
from tqdm import tqdm
import gc
import matplotlib.pyplot as plt
import string

class GCGStyleOptimizerWithPenalty:
    def __init__(self, model, tokenizer, prompt, suffix_init="########",
                 N=100, T=80, k=128, eval_batch_size=32,
                 repeat_lambda=1.0, max_repeat_penalty_coef=10.0):
        """
        参数:
            N: 生成步数（用于计算平均 EOS logit）
            T: 优化迭代次数
            k: 每个位置保留的候选 token 数量
            eval_batch_size: 评估候选时的批次大小（控制显存）
            repeat_lambda: 重复惩罚的权重（与 EOS logit 相加）
            max_repeat_penalty_coef: 最大连续重复惩罚的缩放系数
        """
        self.model = model
        self.tokenizer = tokenizer
        self.device = model.device

        # 编码 prompt 和初始后缀
        self.prompt_tokens = tokenizer.encode(prompt, add_special_tokens=True,
                                               return_tensors='pt').to(self.device)
        suffix_text = suffix_init
        self.suffix_tokens = tokenizer.encode(suffix_text, add_special_tokens=False,
                                               return_tensors='pt').to(self.device)

        self.N = N
        self.T = T
        self.k = k
        self.eval_batch_size = eval_batch_size
        self.repeat_lambda = repeat_lambda
        self.max_repeat_penalty_coef = max_repeat_penalty_coef

        self.vocab_size = model.config.vocab_size
        self.eos_token_id = tokenizer.eos_token_id

        self.loss_history = []          # 记录组合损失
        self.raw_eos_history = []       # 记录原始 EOS logit
        self.repeat_metric_history = [] # 记录平均最大连续重复长度

        # 构建允许的 token 掩码（仅保留 ASCII 可打印字符）
        self.allowed_mask = self._build_allowed_token_mask()
        print(f"允许的 token 数量: {self.allowed_mask.sum().item()} / {self.vocab_size}")

        embedding_dtype = model.get_input_embeddings().weight.dtype
        print(f"初始化完成:")
        print(f"  Prompt tokens: {self.prompt_tokens.shape}")
        print(f"  Suffix tokens: {self.suffix_tokens.shape}")
        print(f"  EOS token ID: {self.eos_token_id}")
        print(f"  Vocab size: {self.vocab_size}")
        print(f"  Embedding dtype: {embedding_dtype}")
        print(f"  N={N}, T={T}, k={k}, eval_batch_size={eval_batch_size}")
        print(f"  repeat_lambda={repeat_lambda}, max_repeat_penalty_coef={max_repeat_penalty_coef}")

    def _build_allowed_token_mask(self):
        """构建允许的 token 掩码：只保留 ASCII 可打印字符"""
        allowed_chars = set(string.printable)  # 包含数字、字母、标点、空格等
        mask = torch.zeros(self.vocab_size, dtype=torch.bool)

        # 遍历所有 token id，解码后检查是否所有字符都在 allowed_chars 中
        for i in range(self.vocab_size):
            token = self.tokenizer.decode([i], clean_up_tokenization_spaces=False)
            # 只允许 token 非空且所有字符都是可打印 ASCII
            if token and all(c in allowed_chars for c in token):
                mask[i] = True
        return mask.to(self.device)

    def compute_eos_loss_and_repeat(self, suffix_tokens_batch):
        """
        批量计算平均 EOS logit 和重复惩罚指标
        返回:
            avg_losses: [batch_size] 平均 EOS logit
            repeat_penalties: [batch_size] 重复惩罚（越大表示重复越严重）
            generated_sequences: [batch_size, N] 生成的 token id 序列（用于分析）
        """
        batch_size = suffix_tokens_batch.shape[0]
        prompt_batch = self.prompt_tokens.repeat(batch_size, 1)
        input_ids = torch.cat([prompt_batch, suffix_tokens_batch], dim=1)

        eos_logits = []
        generated = []  # 存储生成的 token id

        with torch.no_grad():
            current_ids = input_ids
            for step in range(self.N):
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]                # [batch, vocab]
                eos_logits.append(logits[:, self.eos_token_id])  # [batch]

                # 贪婪生成下一步（屏蔽 EOS）
                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_tokens = torch.argmax(masked_logits, dim=-1, keepdim=True)  # [batch, 1]
                generated.append(next_tokens.squeeze(-1))

                current_ids = torch.cat([current_ids, next_tokens], dim=1)

        eos_logits = torch.stack(eos_logits, dim=1)               # [batch, N]
        avg_losses = eos_logits.mean(dim=1)                       # [batch]

        generated = torch.stack(generated, dim=1)                 # [batch, N]

        # 计算重复惩罚：最大连续重复长度 / N，再缩放
        max_repeats = torch.zeros(batch_size, device=self.device)
        for b in range(batch_size):
            seq = generated[b].tolist()
            max_rep = 1
            cur_rep = 1
            for i in range(1, self.N):
                if seq[i] == seq[i-1]:
                    cur_rep += 1
                    max_rep = max(max_rep, cur_rep)
                else:
                    cur_rep = 1
            max_repeats[b] = max_rep

        repeat_penalties = self.max_repeat_penalty_coef * (max_repeats / self.N)  # [batch]

        return avg_losses, repeat_penalties, generated

    def compute_loss_with_grad(self, suffix_tokens):
        """
        计算梯度（基于第一步的 EOS logit，作为整体趋势的代理）
        suffix_tokens: [1, suffix_len]
        返回: (avg_eos_logit, gradients)  gradients: [1, suffix_len, vocab_size]
        """
        embed_weights = self.model.get_input_embeddings().weight
        target_dtype = embed_weights.dtype

        # One-hot relaxation
        suffix_onehot = F.one_hot(suffix_tokens[0], num_classes=self.vocab_size).to(target_dtype)
        suffix_onehot.requires_grad_(True)

        # 计算 suffix embeddings
        suffix_embeds = torch.matmul(suffix_onehot, embed_weights).unsqueeze(0)  # [1, suffix_len, hidden]

        # 获取 prompt embeddings
        with torch.no_grad():
            prompt_embeds = self.model.get_input_embeddings()(self.prompt_tokens)

        # 确保 dtype 一致
        if suffix_embeds.dtype != prompt_embeds.dtype:
            suffix_embeds = suffix_embeds.to(prompt_embeds.dtype)

        # 拼接
        input_embeds = torch.cat([prompt_embeds, suffix_embeds], dim=1)

        # 第一步前向传播
        outputs = self.model(inputs_embeds=input_embeds, use_cache=False)
        first_logits = outputs.logits[:, -1, :]                 # [1, vocab_size]
        first_eos_logit = first_logits[0, self.eos_token_id]

        # 继续生成 N-1 步（无梯度），累加 EOS logit
        with torch.no_grad():
            current_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)
            current_ids = torch.cat([current_ids, torch.argmax(first_logits, dim=-1, keepdim=True)], dim=1)

            total_eos_logit = first_eos_logit.item()

            for step in range(1, self.N):
                outputs = self.model(current_ids, use_cache=False)
                logits = outputs.logits[:, -1, :]
                eos_logit = logits[0, self.eos_token_id]
                total_eos_logit += eos_logit.item()

                masked_logits = logits.clone()
                masked_logits[:, self.eos_token_id] = -float('inf')
                next_token = torch.argmax(masked_logits, dim=-1, keepdim=True)
                current_ids = torch.cat([current_ids, next_token], dim=1)

        # 仅对第一步的 EOS logit 反向传播
        loss = first_eos_logit
        loss.backward()

        gradients = suffix_onehot.grad.unsqueeze(0)              # [1, suffix_len, vocab_size]
        avg_loss = total_eos_logit / self.N

        return avg_loss, gradients

    def get_top_k_candidates_with_filter(self, gradients, suffix_tokens):
        """
        为每个位置选取 top-k 候选 token（基于负梯度），并过滤掉不允许的 token
        """
        suffix_len = suffix_tokens.shape[1]
        candidates = {}
        for pos in range(suffix_len):
            grad = gradients[0, pos, :]                         # [vocab_size]
            # 负梯度越大，替换后越可能降低损失
            top_indices = torch.topk(-grad, self.k).indices     # [k]
            # 只保留允许的 token
            allowed_mask = self.allowed_mask[top_indices]
            filtered = top_indices[allowed_mask].cpu().tolist()
            # 如果过滤后不足，可以保留一些，但通常足够
            if len(filtered) == 0:
                # 保底：取梯度最大的一个允许的 token（即使不在 top-k 中）
                # 简单起见，取整个词汇表中允许的 token 里梯度最大的
                allowed_grads = -grad[self.allowed_mask]
                if allowed_grads.numel() > 0:
                    best_allowed_idx = torch.argmax(allowed_grads)
                    # 找到对应的原始 token id
                    allowed_token_ids = torch.where(self.allowed_mask)[0]
                    filtered = [allowed_token_ids[best_allowed_idx].item()]
                else:
                    filtered = []  # 无候选，后续会跳过该位置
            candidates[pos] = filtered
        return candidates

    def optimize(self):
        print(f"\n开始 GCG 风格优化（带重复惩罚 & token 过滤）")
        print(f"T={self.T}, N={self.N}, k={self.k}, repeat_lambda={self.repeat_lambda}")
        print("=" * 70)

        current_suffix = self.suffix_tokens.clone()

        for iteration in tqdm(range(self.T), desc="优化进度"):
            if iteration % 10 == 0:
                torch.cuda.empty_cache()
                gc.collect()

            # 1. 计算梯度，获取候选
            loss, gradients = self.compute_loss_with_grad(current_suffix)
            candidates = self.get_top_k_candidates_with_filter(gradients, current_suffix)

            suffix_len = current_suffix.shape[1]
            best_suffix = current_suffix.clone()
            best_combined_loss = loss  # 当前组合损失（EOS + 重复惩罚）的占位，初始用纯 EOS 损失，稍后会评估

            # 2. 遍历每个位置及其候选，寻找最佳替换
            for pos in range(suffix_len):
                pos_candidates = candidates[pos]                 # list of token ids
                if not pos_candidates:
                    continue
                num_cand = len(pos_candidates)

                # 分批评估，避免显存溢出
                for i in range(0, num_cand, self.eval_batch_size):
                    batch_cand = pos_candidates[i:i+self.eval_batch_size]
                    batch_size_curr = len(batch_cand)

                    # 构造 batch 后缀
                    cand_suffixes = current_suffix.repeat(batch_size_curr, 1)  # [batch_size_curr, suffix_len]
                    cand_suffixes[:, pos] = torch.tensor(batch_cand, device=self.device)

                    # 评估这批候选：获取 EOS 损失和重复惩罚
                    eos_losses, repeat_penalties, _ = self.compute_eos_loss_and_repeat(cand_suffixes)

                    # 组合损失
                    combined_losses = eos_losses + self.repeat_lambda * repeat_penalties

                    # 找出这批中最好的
                    min_idx = torch.argmin(combined_losses).item()
                    if combined_losses[min_idx] < best_combined_loss:
                        best_combined_loss = combined_losses[min_idx].item()
                        best_suffix = cand_suffixes[min_idx:min_idx+1]

            # 3. 更新当前后缀
            current_suffix = best_suffix
            self.loss_history.append(best_combined_loss)

            # 为了监控，我们重新评估一次当前后缀的原始 EOS 和重复指标
            with torch.no_grad():
                eos_curr, rep_curr, _ = self.compute_eos_loss_and_repeat(current_suffix)
                self.raw_eos_history.append(eos_curr.item())
                self.repeat_metric_history.append(rep_curr.item())

            # 4. 打印进度
            if (iteration + 1) % 10 == 0 or iteration == 0:
                suffix_text = self.tokenizer.decode(current_suffix[0], skip_special_tokens=True)
                print(f"\nIter {iteration+1}/{self.T}:")
                print(f"  组合损失: {best_combined_loss:.4f} (EOS: {eos_curr.item():.4f}, 重复惩罚: {rep_curr.item():.4f})")
                print(f"  当前后缀: '{suffix_text}'")
                if torch.cuda.is_available():
                    print(f"  GPU内存: {torch.cuda.memory_allocated()/1e9:.2f}GB")

        print("\n" + "=" * 70)
        print("优化完成!")
        return current_suffix

    def test_generation(self, suffix_tokens, max_new_tokens=2000):
        """测试生成效果（贪婪解码）"""
        print("\n" + "=" * 70)
        print("测试生成效果:")
        print("=" * 70)

        input_ids = torch.cat([self.prompt_tokens, suffix_tokens], dim=1)

        with torch.no_grad():
            outputs = self.model.generate(
                input_ids,
                max_new_tokens=max_new_tokens,
                do_sample=False,
                pad_token_id=self.tokenizer.eos_token_id,
                use_cache=True
            )

        generated_text = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_tokens = outputs.shape[1] - input_ids.shape[1]

        print(f"\n生成token数: {generated_tokens}")
        # 打印前500字符预览
        print(f"生成预览: {generated_text[:500]}...")

        return generated_text, generated_tokens


# ==================== 主程序 ====================
if __name__ == "__main__":
    # GPU检查与模型加载（同之前代码，略）
    # 此处仅展示如何使用优化器

    # 假设 model, tokenizer 已加载
    prompt = "Please explain quantum computing in simple terms."

    # 基线测试（略）

    optimizer = GCGStyleOptimizerWithPenalty(
        model=model,
        tokenizer=tokenizer,
        prompt=prompt,
        suffix_init="########",
        N=100,
        T=32,
        k=64,
        eval_batch_size=32,
        repeat_lambda=1.0,          # 可调，平衡 EOS 和重复惩罚
        max_repeat_penalty_coef=10.0 # 将最大连续重复长度比例放大10倍，与 EOS logit 量级匹配
    )

    optimized_suffix = optimizer.optimize()
    optimized_text, optimized_length = optimizer.test_generation(optimized_suffix)

    # 绘制损失曲线
    if len(optimizer.loss_history) > 0:
        fig, axes = plt.subplots(2, 1, figsize=(10, 8))
        axes[0].plot(optimizer.loss_history, label='Combined Loss')
        axes[0].set_xlabel('Iteration')
        axes[0].set_ylabel('Loss')
        axes[0].legend()
        axes[0].grid(True)

        axes[1].plot(optimizer.raw_eos_history, label='Avg EOS Logit')
        axes[1].plot(optimizer.repeat_metric_history, label='Repeat Penalty')
        axes[1].set_xlabel('Iteration')
        axes[1].set_ylabel('Value')
        axes[1].legend()
        axes[1].grid(True)

        plt.tight_layout()
        plt.savefig('optimization_with_penalty.png', dpi=150)
        print("\n损失曲线已保存至 optimization_with_penalty.png")
```
